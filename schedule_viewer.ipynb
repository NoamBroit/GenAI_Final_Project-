# Schedule Viewer
מציג את היומן מה-DB בצורת DataFrame

```python
import sqlite3
import pandas as pd
import os

DB_PATH = os.path.join(os.getcwd(), "tech.db")
conn = sqlite3.connect(DB_PATH)

df = pd.read_sql_query("SELECT * FROM Schedule ORDER BY date, time", conn)
conn.close()

print(f"Total rows: {len(df)}")
df.head(20)
```

## סלוטים פנויים בלבד

```python
from datetime import date

df_available = df[
    (df["available"] == 1) &
    (df["date"] >= date.today().isoformat())
].reset_index(drop=True)

print(f"Available slots from today: {len(df_available)}")
df_available.head(20)
```

## פילטור לפי פוזיציה

```python
position = "Python Dev"  # שנה לפי הצורך: Python Dev, Sql Dev, Analyst, ML

df_position = df_available[df_available["position"] == position].reset_index(drop=True)

print(f"Available slots for '{position}': {len(df_position)}")
df_position.head(20)
```

```python
print("=" * 40)
print("Schedule Statistics")
print("=" * 40)
print(f"Total slots:     {len(df)}")
print(f"Available:       {len(df[df['available'] == 1])}")
print(f"Taken:           {len(df[df['available'] == 0])}")
print()
print("Available by position:")
print(df[df['available'] == 1].groupby('position').size().to_string())
```
